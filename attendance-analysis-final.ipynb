# Binding different subject attendance details into one 

```r
data = cbind(Internet_Of_Things[,1:4] , Machine_Learning[,4] , 
             Data_Science[,4] , Embedded_System[,4] , Block_Chain[,4])
colnames(data)[5] ="No_of_classes_present_ml"
colnames(data)[6] ="No_of_classes_present_ds"
colnames(data)[7] ="No_of_classes_present_es"
colnames(data)[8] ="No_of_classes_present_bc"
data
```

# Calculating attendance Percentage for each student for each subject

```r
iot = c()
ml = c()
ds = c()
es = c()
bc = c()
```

```r
for (i in 1:50)
{
    x = data %>% slice(i)
    iot = append(iot , round(((x[["No_of_classes_present_iot"]]/45)*100),digits=2))
    ml = append(ml , round(((x[["No_of_classes_present_ml"]]/45)*100),digits=2))
    ds = append(ds , round(((x[["No_of_classes_present_ds"]]/45)*100),digits=2))
    es = append(es , round(((x[["No_of_classes_present_es"]]/45)*100),digits=2))
    bc = append(bc , round(((x[["No_of_classes_present_bc"]]/45)*100),digits=2))
}
```

```r
data$IOT_Present_Percent = iot
data$ML_Present_Percent = ml
data$DS_Present_Percent = ds
data$ES_Present_Percent = es
data$BC_Present_Percent = bc
```

```r
total_percent = c()

for (i in 1:50)
{
    tot = 0 
    x = data %>% slice(i)
    tot = tot + x[["IOT_Present_Percent"]] 
    tot = tot + x[["ML_Present_Percent"]] 
    tot = tot + x[["DS_Present_Percent"]] 
    tot = tot + x[["ES_Present_Percent"]] 
    tot = tot + x[["BC_Present_Percent"]] 
    total_percent = append(total_percent , round((tot/5),digits=2))
}
data$Total_Present_Percent = total_percent
```

# Individual Students Attendance Analysis

```r
input <- function()
{
    reg_no = readline(prompt("Enter the Register Number : "))
    reg_no = as.integer(reg_no)
    return (reg_no)
}    
```

```r
reg_no = input()
```

```r
fetch <- function()
{
    student_details = c()
    for ( i in 1:50)
    {
        x = data %>% slice(i)
        if (x[["Register_Number"]] == reg_no)
        {
            student_details = c(x[["IOT_Present_Percent"]] , 
                                x[["ML_Present_Percent"]],x[["DS_Present_Percent"]],
                                x[["ES_Present_Percent"]],x[["BC_Present_Percent"]])
            print(student_details)
        }    
    }
    return (student_details)
}
```

```r
student_details = fetch()
student_details
```

```r
if (length(student_details) == 0)
{
    print("Incorrect Register Number ")
    print("Enter the correct Register Number")
    reg_no = input()
    student_details = fetch()
    student_details
}
```

```r
per_student_data = data.frame("Subjects" = subjects, "Percentage" = student_details)
per_student_data
```

```r
str = paste("   Attendance Percentage of ",as.character(reg_no))
ggplot(per_student_data, aes(Subjects, Percentage)) + 
geom_bar(stat="identity", 
         fill="Yellow", 
         color="Maroon") + 
ggtitle(str , subtitle = "") + 
theme(legend.position = "none", 
      plot.title = element_text(hjust = 0.5), 
      plot.subtitle = element_text(hjust = 0.5)) + 
scale_y_continuous(labels=comma) +
geom_text(aes(label=Percentage))
```
